```python


import pandas as pd
df = pd.read_csv("creditcard.csv")


df_mean = df.fillna(df.mean())
df_median = df.fillna(df.median())
df_mode = df.fillna(df.mode().iloc[0])
df_dropna = df.dropna()

print("DataFrame con valores nulos reemplazados por la media:")
print(df_mean.head())

print("\nDataFrame con valores nulos reemplazados por la mediana:")
print(df_median.head())

print("\nDataFrame con valores nulos reemplazados por la moda:")
print(df_mode.head())

print("\nDataFrame con filas con valores nulos eliminadas:")
print(df_dropna.head())
```
